```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
data = pd.read_csv('data/people.csv')
```

```python
data.columns
```

```python
# count the number of unique defendants in the data set
data['id'].count()
```

```python
# plot the distribution of decile scores
sns.distplot(data['decile_score'], hist=True, kde=False,hist_kws={'edgecolor':'black'})

# there are ~40 records with decile score = -1; we probably want to remove these from the sample.
```

```python
# plot the number of defendants of each race
race = data['race'].value_counts()
race.plot(kind='bar')
```

```python
# plot the number of defendants of each sex
sex = data['sex'].value_counts()
sex.plot(kind='bar')
```

```python
# plot the distribution of priors counts
sns.distplot(data['priors_count'], hist=False, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1})
```
